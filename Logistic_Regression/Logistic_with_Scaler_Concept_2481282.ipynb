```python
import pandas as pd
import seaborn as sns
```

```python
d1=pd.read_csv('Facebook_Ads_2.csv',encoding='ISO-8859-1')
d1
```

```python
d1.drop(['Names','emails','Country'],axis=1,inplace=True)
```

```python
X=d1.drop('Clicked',axis=1).values
y=d1['Clicked'].values
```

In This Case Too Much Float Value , We have to take one standard which will applied throuhout the dataset to maintain proper order & standard decimal upto a certain digit.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler() # Just a variable otherwise have to use StandardScaler full word
```

```python
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.28)
```

```python
from sklearn.linear_model import LogisticRegression
logisr=LogisticRegression()
```

```python
logisr.fit(X_train,y_train)
```

```python
y_predict=logisr.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```
