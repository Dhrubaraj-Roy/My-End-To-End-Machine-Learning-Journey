```python
sns.displot(df,x='total_bill',hue='sex')
# hue is the built in function clubed or associated with seaborn 
# which helps user to work with object columns
```

```python
sns.displot(df,x='total_bill',hue='smoker')
```

```python
sns.displot(df,x='total_bill',y='tip')
```

```python
sns.catplot(x='sex',y='tip',data=df,hue='smoker')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
df.head(5)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,hue='sex')
```

In this case Due to Close Nature of Each Point, All Point Plotted properly but in normal view it will start display in overlap way-beacuse mostly we are using circle to plot a point-two circle can bemerged or overlap together very easily,if we can change the geometric style of plotting of the points then issue can beresolved properly.

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
#reg=Regression
```
